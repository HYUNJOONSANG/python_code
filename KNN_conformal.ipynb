3. Implement the Nearest Neighbour method (i.e., the 1 Nearest Neighbours
method). Run it on both datasets: train it on the training set and compute
the number of errors it makes on the test set and the test error rate (the
ratio of the number of errors to the size of the test set).



```python
from collections import Counter # To obtain mode, from 'collection' imported to Counter.

#function 'test_dist' is that calculates a distance 
# between predicted test set and predicted train set, using 'np.linalg.norm' 

def test_dist(test_set, training_set):
    return np.linalg.norm(test_set - training_set)


# function 'neighboursFunc' is that gathers value of each calculated distance 
# and then sort from minimum value to maximum value. Thus, this function provide predicted data.

def neighboursFunc(training_set, label_training_set, test_distance, k):
    list_dist = []
    for index in range(len(training_set)):
        dist = test_dist(test_distance, training_set[index])
        list_dist.append((training_set[index], dist,label_training_set[index]))
    
    testing = np.shape(list_dist)
    newlist = sortingFunc(list_dist)
    neighbours = newlist[:k]
    return neighbours

# function 'sortingFunc' is that sort values from min to max.

def sortingFunc(list_dist):
    for i in range(len(list_dist)-1,0,-1):
        for j in range(i):
            if list_dist[j][1] > list_dist[j+1][1]:
                temp = list_dist[j+1]
                list_dist[j+1] = list_dist[j]
                list_dist[j] = temp
    return list_dist

# function 'accuracyFunc' is that compares predicted labels with true labels,
# and then calculates a percentage of accuracy and number of accuracy.

def accuracyFunc(voteData, testData):
    number_of_accuracy = 0
    for i in range(len(voteData)):
        if voteData[i] == testData[i]:
            number_of_accuracy += 1
    rate_of_accuracy = (number_of_accuracy/len(y_test)) * 100
    new_data = np.zeros(2)
    new_data[0] = rate_of_accuracy
    new_data[1] = number_of_accuracy
    return new_data

# when K value is more than 1, we have to vote which value is frequently used.
# Thus, this function 'geVoteValue' supports our decision. Using function 'most_common', we are able to obtain most frequently used value.
# I referred to 'https://docs.python.org/2/library/collections.html'

def getvoteValue(neighbours):

    voteValue = []

    for index in range(len(neighbours)):
        voteValue.append(neighbours[index][2])

    counts = Counter(voteValue)
    oldneighbours = counts.most_common(1) # In order to get mode, function'most_common' was used 

    return  oldneighbours[0][0]
```

load_iris's Nearest Neighbour algorithm using K=1

```python
voteData = np.zeros(len(y_test))

for i in range(len(X_test)):
    neighbours = neighboursFunc (X_train, y_train, X_test[i], 1) # Using neighboursFunc, it is sent X_train, y_train and K value. 
                                                                 # In here K is 1
    voteData[i] = neighbours[0][2]

number_of_error = 0 
number_of_error = len(y_test) - accuracyFunc(voteData,y_test)[1]

print('percentage of accuracy: ', round((accuracyFunc(voteData,y_test)[0]),4),'%')
print('rate of error of test set: ',round((100 - accuracyFunc(voteData,y_test)[0]),4),'%')
print('number of error: ', int(number_of_error))
```
